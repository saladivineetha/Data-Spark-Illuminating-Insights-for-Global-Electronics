# MySQL Connection

```python
import pymysql
import warnings
warnings.filterwarnings('ignore')
```

```python
# Establish connection to MySQL
myconnection = pymysql.connect(host='localhost', user='root', password='root', database='Global_electronics')
my_cursor = myconnection.cursor()
my_cursor.execute("CREATE DATABASE IF NOT EXISTS GLOBAL_ELECTRONICS")
```

# Inserting Customers_data 

```python
my_cursor.execute('''CREATE TABLE IF NOT EXISTS customer_data(
                  CustomerKey INT,
                  Gender VARCHAR(255),
                  City VARCHAR(255),
                  State VARCHAR(255),
                  Country VARCHAR(255),
                  Continent VARCHAR(255)
                 
                  )''')
print("Table Created successfully")
```

```python
insert_query = '''INSERT INTO customer_data(
                        CustomerKey,
                        Gender,
                        City,
                        State,
                        Country,
                        Continent)
                        VALUES(%s,%s,%s,%s,%s,%s)'''
data=customers_data.values.tolist()
my_cursor.executemany(insert_query,data)
myconnection.commit()
print("Inserted successfully")
```
